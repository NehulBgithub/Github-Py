```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate Data
# Generating random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)


```

```python
data
```

```python
# Step 2: Create Histograms
# Plotting histogram using seaborn
# sns.set(style="whitegrid")
plt.figure(figsize=(6, 6))
sns.histplot(data, bins=30,kde=True, color='blue')

# Adding title and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

```python
# Step 1: Generate Data
# Categories and their corresponding values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 30, 45, 10]
```

```python
# Step 2: Create Pie Chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('Pie Chart of Categories')
plt.show()
```

```python
# Step 3: Create Bar Chart using Seaborn
plt.figure(figsize=(6, 6))
sns.barplot(x=categories, y=values, palette="pastel")
plt.title('Bar Chart of Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(100)  # 100 random points for x-axis
y = np.random.rand(100)  # 100 random points for y-axis
```

```python
# Step 2: Create Scatter Plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x, y=y, color='blue', marker='*')

# Adding title and labels
plt.title('Scatter Plot of Random Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the plot
plt.show()
```

```python
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)

# Create a regression plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x=x, y=y, color='blue')

# Adding title and labels
plt.title('Regression Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
import pandas as pd

# Generate some multivariate data
data = sns.load_dataset("iris")  # Example dataset

# Create a pair plot
sns.set(style="whitegrid")
sns.pairplot(data, hue="species", palette="husl")

plt.show()
```

```python
# Generate some data
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # Sine wave

# Create a line chart
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, color='blue')

# Adding title and labels
plt.title('Line Chart of Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

```python
# Generate some data
categories = ['Category A', 'Category B', 'Category C']
values1 = [10, 20, 30]
values2 = [20, 25, 15]
values3 = [15, 18, 22]

# Create a DataFrame
df = pd.DataFrame({
    'Categories': categories,
    'Series 1': values1,
    'Series 2': values2,
    'Series 3': values3
})

# Plotting a stacked bar chart
df.set_index('Categories').plot(kind='bar', stacked=True, color=['blue', 'orange', 'green'], figsize=(10, 6))

# Adding title and labels
plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```
