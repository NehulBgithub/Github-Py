#### Invalid order IDs
Question 4: How many invalid order IDs are there in the data frame. An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of valid order order IDs in the data frame.
    
    a)6
    b)7
    c)8
    d)9


```python
#your code here
import pandas as pd
def is_valid_order_id(order_id):
    parts = order_id.split('-')
    if len(parts) == 3 and parts[0] == 'AZ' and parts[1] == '2011' and len(parts[2]) == 7 and parts[2].isdigit():
        return True
    return False

# Apply the function to check validity and count valid and invalid IDs
df['is_valid'] = df['Order ID'].apply(is_valid_order_id)

# Count the number of valid and invalid Order IDs
num_valid_order_ids = df['is_valid'].sum()
num_invalid_order_ids = len(df) - num_valid_order_ids

num_invalid_order_ids, num_valid_order_ids

```

```python
df['Order no']=df['Order ID'].str.split('-')

df['Order no']=df['Order no'].apply(lambda x: x[2])

df['no_of_digits']=df['Order no'].apply(lambda x: len(x))

df['no_of_digits'].value_counts()
```
